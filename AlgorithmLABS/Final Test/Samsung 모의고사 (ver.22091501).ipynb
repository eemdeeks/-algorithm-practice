```python
n = int(input())
for i in range(1,n+1):
    answer = 0
    m = int(input())
    ls = []
    for j in range(m):
        ls.append(list(map(int,input().split())))
    for j in range(1,m-1):
        for k in range(1,m-1):
            result = set()
            result.update(ls[j])
            for l in range(m):
                result.add(ls[l][k])
            if answer < sum(result):
                answer = sum(result)
    print("#%d"%i, answer)
```
