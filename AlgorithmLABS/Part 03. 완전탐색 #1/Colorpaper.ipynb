```python
ls = []
for i in range(102):
  ls.append([])
  for j in range(102):
    ls[i].append(0)

n = int(input())
for i in range(1,n+1):
  a = list(map(int,input().split()))
  for j in range(a[0],a[0]+a[2]):
    for k in range(a[1],a[1]+a[3]):
      ls[j][k] = i
for i in range(1,n+1):
  sum = 0
  for j in ls:
    for k in j:
      if k == i:
        sum += 1
  print(sum)  
```
