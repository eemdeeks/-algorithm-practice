```python
a = list(map(int,input().split()))
b = list(map(int,input().split()))
aW = 0
bW = 0
for i in range(10):
  if a[i] > b[i]:
    aW += 1
  elif b[i] > a[i]:
    bW += 1
if aW > bW:
  print("A")
elif bW > aW:
  print("B")
else:
  print("D")
```
