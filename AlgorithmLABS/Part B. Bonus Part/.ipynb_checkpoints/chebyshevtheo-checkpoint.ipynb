```python
### 에라토스테네스의 체 이용해서.
ls = []
for i in range(123456*2+1):
    ls.append(True)
for i in range(2,123456*2+1):
    if ls[i]:
        for j in range(i+1,123456*2+1):
            if j%i == 0:
                ls[j]=False
for i in range(2):
    ls[i] = False
while(True):
    result = 0
    n = int(input())
    for i in range(n+1,2*n+1):
        if ls[i]:
            result+=1
    print(result)

        
```
