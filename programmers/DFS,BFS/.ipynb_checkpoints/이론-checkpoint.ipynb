## DFS

DFS          .
DFS  ( ) ,     .
1.        .
2.               .         .
3.   2        .



- [step 0]  . (  :    )  
  : 1
- [step 1]   '1'     .  
- [step 2]    '1'     '2','3','8' .    
     '2'     .  
- [step 3]    '2'     '7' .    
 '7'      .
- [step 4]    '7'     '6','8' .  
     '6'    .  
- [step 5]    '7'     '8' .  
 '8'      .  

       (  )  .  
 : 1 -> 2 -> 7 -> 6 -> 8 -> 3 -> 4 -> 5

```python
#DFS  
def dfs(graph, v, visited):
    #    
    visited[v] = True
    print(v, end='')
    #       
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i ,visited)
        
#      (2 ) 
# index 0    (     )
graph = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

#      (1 )
visited = [False] * 9

#  DFS  
dfs(graph, 1, visited)
```
